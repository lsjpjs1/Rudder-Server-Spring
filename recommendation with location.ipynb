```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

# Read Data


```python
locations = pd.read_excel('/content/drive/MyDrive/rudder/data/rudder-location.xlsx')
partyInfo = pd.read_excel('/content/drive/MyDrive/rudder/data/rudder-partyInfo.xlsx')
ratings = pd.read_excel('/content/drive/MyDrive/rudder/data/rudder-rating.xlsx')
```

```python
locations.to_csv('/content/drive/MyDrive/rudder/data/rudder-location.csv', index=None, header=True)
partyInfo.to_csv('/content/drive/MyDrive/rudder/data/rudder-partyInfo.csv', index=None, header=True)
ratings.to_csv('/content/drive/MyDrive/rudder/data/rudder-rating.csv', index=None, header=True)
```

```python
partyInfo
```

```python
ratings.sample()
```

```python
ratings = pd.merge(ratings, partyInfo, on='partyId', how='left')
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(ratings, random_state=42, test_size=0.1)
```

```python
train.shape
```

```python
loca_cols = ratings.columns[4:]
loca_cols
train.sample()
```

```python
for loca in loca_cols:
  train[loca] = train[loca] * train['rating']
```

```python
train.sample()
```

# User Profile

```python
user_profile = train.groupby('userId')[loca_cols].mean()
```

```python
user_profile
```

# Sample Predict


```python
sample = test.iloc[5]
sample_user = sample['userId']
sample_user_profile = user_profile.loc[sample_user]
sample_user_profile
```

```python
sample[loca_cols]
```

```python
(sample_user_profile * sample[loca_cols]).mean()
```

# 전체 데이터 predict

```python
from tqdm import tqdm_notebook
```

```python
predict = []
for idx, row in tqdm_notebook(test.iterrows()):
  user = row['userId']
  if user in user_profile.index:
    predict.append((user_profile.loc[user] * row[loca_cols]).mean())
  else: 
    predict.append(np.nan)
predict
```

```python
test['predict'] = predict
test.head()
test.loc[test['predict'].isnull(), 'predict'] = train['rating'].mean()
test.head()
```

# model validation

```python
from sklearn.metrics import mean_squared_error
```

```python
mean = mean_squared_error(test['rating'], test['predict'])
rmse = np.sqrt(mean)
rmse
```
